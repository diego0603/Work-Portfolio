### Marco teórico

#### Optimización Bayesiana

**Definición y contexto:** La optimización bayesiana es un método probabilístico para optimizar funciones objetivo que son caras de evaluar. Este enfoque es utilizado en problemas donde las evaluaciones experimentales son costosas o complejas (Shahriari et al., 2016).

**Elementos clave:**

+ **Modelo probabilístico:**  Los Gaussian Processes (GP) se utilizan como modelos que ajustan los datos experimentales y predicen resultados en regiones no exploradas.

+ **Funciones de adquisición:** Estas guían la selección de nuevos puntos de prueba equilibrando explotación (probar cerca de buenos resultados conocidos) y exploración (investigar nuevas regiones).

**Ventajas:**

+ Reducción del número de experimentos necesarios.

+ Capacidad de optimizar problemas con comportamientos no lineales.

**Aplicaciones:** La optimización bayesiana se aplica en ingeniería de materiales, optimización de procesos químicos, y diseño de experimentos (Frazier, 2018).

#### El Proceso de Sputtering (Pulverización Catódica)
**Fundamentos**: El sputtering es una técnica de deposición física de vapor (PVD) utilizada para fabricar películas delgadas (Smith, 1995).

**Proceso básico:**

+ Un gas (generalmente argón) es ionizado formando un plasma.

+ Los iones bombardean un blanco (target) de material sólido, liberando átomos o moléculas.

+ Estos átomos se depositan sobre un sustrato formando una película delgada.

**Parámetros clave:**

+ Potencia (DC o RF): Afecta la energía del plasma y la tasa de deposición.

+ Presión del gas: Influye en la colisión de partículas y el espesor uniforme.

+ Flujo de gases reactivos: Controla la composición química de la película.

**Desafíos:** Optimizar estos parámetros para obtener películas con propiedades deseadas (espesor, uniformidad, adhesión, etc.) (Ohring, 2002).

#### Películas Delgadas
**Definición:** Materiales con espesores en el rango nanométrico a micrométrico depositados sobre un sustrato (Ohring, 1995).

**Propiedades y aplicaciones:**

+ Propiedades ópticas: Antirreflectantes, filtros de color.

+ Propiedades eléctricas: Conductividad, semiconductores.

+ Propiedades mecánicas: Resistencia al desgaste, dureza.

+ Aplicaciones: Utilizadas en electrónica, óptica, energías renovables y recubrimientos protectores (Smith, 1995).

**Impacto de los parámetros de sputtering:** Los parámetros del proceso afectan características como la estructura cristalina, la composición química y las propiedades funcionales (Ohring, 2002).

#### Relación entre Optimización Bayesiana, Sputtering y Películas Delgadas
**Optimización Bayesiana:** La optimización bayesiana puede abordar los desafíos del diseño experimental en sputtering, reduciendo el tiempo y los costos para obtener películas con propiedades específicas (Brochu, Cora, & De Freitas, 2010).

**Ejemplos de objetivos:**

+ Maximizar la uniformidad del espesor.

+ Minimizar defectos en la película.

+ Ajustar propiedades ópticas o eléctricas.

### Descripción del modelo

El modelo que se implementó empleó un enfoque de Optimización Bayesiana basado en un Gaussian Process Regressor (GPR). Este modelo probabilístico ajusta una función objetivo a partir de los datos experimentales iniciales, lo que permite predecir el comportamiento de la señal objetivo en áreas no exploradas del espacio de parámetros. La optimización bayesiana resulta especialmente útil cuando las evaluaciones experimentales son costosas o complejas, como es el caso en el proceso de sputtering.

#### Criterio de Optimización

El criterio de optimización se basa en minimizar la diferencia absoluta entre la señal calculada y una señal objetivo predefinida, que en este caso es 50. El modelo busca encontrar las combinaciones de parámetros de entrada (potencia_1 y potencia_2) que produzcan la señal más cercana al valor deseado.

**criterio formal**

$ min|señal(potencia_1, potencia_2) - Señal Objetivo|$

#### Variables de Control

**Potencia 1 (potencia_1):**

+ Representa un parámetro ajustable relacionado con la intensidad de la fuente de energía.
+ Intervalo de valores: [10, 100].
  
**Potencia 2 (potencia_2):**

+ Representa un parámetro de control que afecta las condiciones del flujo de gases o presión de trabajo.
+ Intervalo de valores: [10, 100].

#### Función de costo

$Costo = |señal(potencia_1, potencia_2) - Señal Objetivo|$

#### Rango de Variables de Control:

+ $10 \leq potencia_1 \leq 100$
+ $10 \leq potencia_2 \leq 100$
  
Estas restricciones aseguran que las combinaciones de parámetros probadas sean físicamente viables y relevantes para el proceso experimental.

```python
#Código creado por Sebastián Pérez Gascon
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF
import numpy as np
import matplotlib.pyplot as plt

def funcion_costo(x):
    potencia_1, potencia_2 = x[:, 0], x[:, 1]
    señal = 0.8 * potencia_1 + 0.2 * potencia_2 + np.sin(potencia_1) - np.cos(potencia_2)
    señal_objetivo = 50
    return abs(señal - señal_objetivo)


class OptimizacionSputtering:
    def __init__(self, num_iteraciones=20):
        self.num_iteraciones = num_iteraciones
        self.kernel = 1.0 * RBF(length_scale=1.0)
        self.gp = GaussianProcessRegressor(kernel=self.kernel, n_restarts_optimizer=10)
        
        self.limites = [10, 100]
    
    def optimizar(self):

        X1 = np.random.uniform(self.limites[0], self.limites[1], (10, 2))
        y1 = funcion_costo(X1)
        
        X_historia = X1
        y_historia = y1
    
        for _ in range(self.num_iteraciones):
            self.gp.fit(X1, y1)
            

            x1 = np.linspace(self.limites[0], self.limites[1], 100)
            x2 = np.linspace(self.limites[0], self.limites[1], 100)
            x1_grid, x2_grid = np.meshgrid(x1, x2)
            x_pred = np.c_[x1_grid.ravel(), x2_grid.ravel()]
        
            y_pred, y_std = self.gp.predict(x_pred, return_std=True)

            nuevo_punto = x_pred[np.argmax(y_std)]
            nuevo_punto = nuevo_punto.reshape(1, -1)
    
            nuevo_y = funcion_costo(nuevo_punto)
        
            X1 = np.vstack([X1, nuevo_punto])
            y1 = np.append(y1, nuevo_y)
            
            X_historia = np.vstack([X_historia, nuevo_punto])
            y_historia = np.append(y_historia, nuevo_y)
        
        mejor_indice = np.argmin(y1)
        return X1[mejor_indice], y1[mejor_indice], X_historia, y_historia

    def visualizar_resultados(self, X_historia, y_historia)
        plt.figure(figsize=(15, 5))
        
        plt.subplot(121)
        x1 = np.linspace(self.limites[0], self.limites[1], 100)
        x2 = np.linspace(self.limites[0], self.limites[1], 100)
        x1_grid, x2_grid = np.meshgrid(x1, x2)
        x_pred = np.c_[x1_grid.ravel(), x2_grid.ravel()]
        
        y_pred, y_std = self.gp.predict(x_pred, return_std=True)
        
        plt.contourf(x1_grid, x2_grid, y_pred.reshape(x1_grid.shape), 
                     levels=50, cmap="viridis", alpha=0.8)
        plt.colorbar(label="Predicción de diferencia de señal")
        
        plt.scatter(X_historia[:, 0], X_historia[:, 1], c=y_historia, s=50, cmap="Reds", edgecolor="k", label="Puntos muestreados")
        plt.xlabel("Potencia 1")
        plt.ylabel("Potencia 2")
        plt.title("Mapa de Predicción")
        plt.legend()
        
        
        plt.subplot(122)
        plt.plot(np.minimum.accumulate(y_historia))
        plt.title("Convergencia del Valor Mínimo")
        plt.xlabel("Iteración")
        plt.ylabel("Mejor Valor Encontrado")
        
        plt.tight_layout()
        plt.show()

optimizador = OptimizacionSputtering(num_iteraciones=20)
mejor_punto, mejor_valor, X_historia, y_historia = optimizador.optimizar()

print("Mejor punto encontrado:")
print(f"Potencia 1: {mejor_punto[0]}")
print(f"Potencia 2: {mejor_punto[1]}")
print(f"Valor en ese punto: {mejor_valor}")

optimizador.visualizar_resultados(X_historia, y_historia)
```

```python
#Código creado por Diego Yael Islas Santoyo
#Código a evaluar
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF
import numpy as np
import matplotlib.pyplot as plt

def parametros(x):
    potencia_1, potencia_2 = x[:, 0], x[:, 1]
    señal = 0.8 * potencia_1 + 0.2 * potencia_2 + np.sin(potencia_1) - np.cos(potencia_2)
    señal_objetivo = 50
    return abs(señal - señal_objetivo), señal  

# Generación de datos iniciales
X1 = np.random.uniform(10, 100, (10, 2))
errores, señales = parametros(X1)

# Ajuste del modelo GP
kernel = 1.0 * RBF(length_scale=1)
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10)
gp.fit(X1, errores)

# Predicción en una cuadrícula
x1 = np.linspace(10, 100, 100)
x2 = np.linspace(10, 100, 100)
x1_grid, x2_grid = np.meshgrid(x1, x2)
x_pred = np.c_[x1_grid.ravel(), x2_grid.ravel()]
y_pred, y_std = gp.predict(x_pred, return_std=True)

# Potencia obtenida y error absoluto
potencia_minima = X1[np.argmin(errores)]  
error_minimo = np.min(errores)
potencia_minima_valor = señales[np.argmin(errores)]

# Gráfico principal
plt.figure(figsize=(10, 8))
plt.contourf(x1_grid, x2_grid, y_pred.reshape(x1_grid.shape), levels=50, cmap="viridis", alpha=0.8)
plt.colorbar(label="Predicción de diferencia de señal")
plt.scatter(X1[:, 0], X1[:, 1], c=errores, s=50, cmap="Reds", edgecolor="k", label="Datos de entrenamiento")
plt.xlabel("Potencia 1")
plt.ylabel("Potencia 2")
plt.legend()
plt.title("Predicción de la diferencia de señal (GP)")
plt.grid()
plt.show()

# Gráfico de iteraciones vs error
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(errores) + 1), errores, marker="o", label="Error absoluto")
plt.axhline(y=error_minimo, color="r", linestyle="--", label=f"Mínimo error: {error_minimo:.2f}")
plt.xlabel("Iteración")
plt.ylabel("Error absoluto")
plt.title("Iteraciones y Error Absoluto")
plt.legend()
plt.grid()
plt.show()

# Resultados obtenidos
print("Potencia obtenida (coordenadas):", potencia_minima)
print("Señal más cercana al objetivo:", potencia_minima_valor)
print("Error absoluto mínimo:", error_minimo)
```
